```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
X, y, cols = data.data, data.target, data.feature_names
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(data=X, index=range(X.shape[0]), columns=cols)
df['diagnosis'] = y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_visualized = pd.concat([df.loc[:, 'mean radius':'mean fractal dimension'], df['diagnosis']], axis=1)
g = sns.PairGrid(df_visualized, hue='diagnosis')
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
```
